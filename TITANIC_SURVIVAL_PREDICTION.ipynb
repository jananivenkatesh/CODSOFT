```python
#splitting the data
Y=df.Survived.copy()
X=df.drop(['Survived'],axis=1)
```

```python
#drop the attribbuted which does not have a stronger correlation to affect the outcome
X.drop(['Cabin', 'Ticket', 'Name', 'PassengerId'], axis=1, inplace=True)
```

```python
X.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
df['Sex']=labelencoder.fit_transform(df['Sex'])
df.head()
```

```python
#as per prior results we have null values in the age component , we are replacing the null value with mean
X.Age.fillna(X.Age.mean(), inplace=True)
```

```python
#check any null values present
X.isnull().values.any()
```

```python
#import libraries for model building
from sklearn.model_selection import train_test_split
#splitting the data to train our model and test its accuracy
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, random_state=7)
```

```python
#import regression libraries for model building
from sklearn.linear_model import LogisticRegression
#initialize it to a name
model=LogisticRegression()
#fitting data to the model
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
#predict in the model
Y_pred_test=model.predict(X_test)
Y_pred_test
```
